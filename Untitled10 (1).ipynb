```python
gray = cv.cvtColor(image_copy, cv.COLOR_RGB2GRAY)
gray = np.float32(gray)

# Detect corners 
dst = cv.cornerHarris(gray, 2, 3, 0.02)

# Dilate corner image to enhance corner points
dst = cv.dilate(dst,None)

plt.imshow(dst, cmap='gray')
```

```python
import cv2
import numpy
from PIL import Image
from skimage.filters.thresholding import threshold_otsu
#Segmentation 171
# Opening the image and converting it to grayscale
a = Image.open(path).convert('L')
a = numpy.asarray(image_copy)
thresh = threshold_otsu(a)
# Pixels with intensity greater than the
# "threshold" are kept.
b = 255*(a > thresh)
# Saving the image.
#cv2.imwrite(r"C:\Users\Pratyush\Downloads", b)
plt.imshow(b)
```

```python
thresh = 0.01*dst.max()
x_max=float('-inf')
y_max=float('-inf')
x_min=float('inf')
y_min=float('inf')
# Create an image copy to draw corners on
corner_image = np.copy(b)
count=[]
# Iterate through all the corners and draw them on the image (if they pass the threshold)
for j in range(0, dst.shape[0]):
    for i in range(0, dst.shape[1]):
        if(dst[j,i] > thresh):
            
            # image, center pt, radius, color, thickness
            cv.circle( corner_image, (i, j), 1, (0,255,0), 10)
            print(i,j)
            if i>x_max:
                x_max=i
            if j>y_max:
                y_max=j
            if i<x_min:
                x_min=i
            if j<y_min:
                y_min=j
            count.append([i,j])
plt.imshow(corner_image)
```
